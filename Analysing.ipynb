```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle
import numpy as np
def load_pickle(file):
    with open(file+".pckl", mode='rb') as f:
        return pickle.load(f)
    
def save_pickle(filename,obj):
    print("saved as", filename + ".pckl")
    with open(filename+".pckl", mode='wb') as f:
        pickle.dump(obj, f)   

#%matplotlib notebook
table=load_pickle("scaletable2")
table=table["table"]

Xdata=np.array(table)[1:,0].astype(float)
print(Xdata)
Ydata=np.array(table)[1:,1].astype(float)
print(Ydata)
Zdata=np.array(table)[1:,2].astype(float)
print(Zdata)
Zdata2=np.array(table)[1:,3].astype(float)
print(Zdata2)

#%matplotlib inline
%matplotlib notebook
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
surf = ax.scatter(Xdata, Ydata,Zdata,'r',s=7)
surf2 = ax.scatter(Xdata, Ydata,Zdata2,'g',s=7)
#https://stackoverflow.com/questions/29549905/pylab-3d-scatter-plots-with-2d-projections-of-plotted-data
```

```python
import matplotlib.patches as mpatches
%matplotlib inline
f, (ax1, ax2) = plt.subplots(2, figsize=(10,10))
f.subplots_adjust(hspace = .4, wspace=.2)
line1=ax1.plot(Xdata, Zdata,'b.')
line2=ax1.plot(Xdata, Zdata2,'r.')
red_patch = mpatches.Patch(color='red', label='False Positive')
blue_patch = mpatches.Patch(color='blue', label='True Positive')
ax1.legend(handles=[blue_patch,red_patch])
ax2.plot(Ydata, Zdata,'.b')
ax2.plot(Ydata, Zdata2, 'r.')
ax2.legend(handles=[blue_patch,red_patch])
#ax1.set_title('y-start', fontsize=20)
#ax2.set_title('scale', fontsize=20)
ax1.set_xlabel('y-start')
ax1.set_ylabel('% of region')
ax2.set_xlabel('scale')
ax2.set_ylabel('% of region')

#ax1.imshow(carExampleImage)
#ax1.set_title('Car', fontsize=20)
#ax2.imshow(carhogimage, cmap='gray')
#ax2.set_title('Car HOG', fontsize=20)
#ax3.imshow(notExampleImage)
#ax3.set_title('Not Car', fontsize=20)
#ax4.imshow(nothogimage, cmap='gray')
#ax4.set_title('Not Car HOG', fontsize=20)

#

```

```python
from matplotlib.mlab import griddata
xi = np.linspace(Xdata.min()-10, Xdata.max()+10, 1000)
yi = np.linspace(Ydata.min()-1, Ydata.max()-1, 1000)
zi = griddata(Xdata, Ydata, Zdata, xi, yi, interp='linear')
%matplotlib notebook
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(xi, yi,zi)
```
